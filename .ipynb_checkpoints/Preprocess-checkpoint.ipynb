```python
import os
```

```python
def get_pathway_sentences(file_path):
    sentences = ''
    with open(file_path,'r') as reader:
        for i,line in enumerate(reader):
            if i==0:
                sentences+= line.strip()
            if i>=1:
                sentences+= ' '+line[2:].strip()
    return sentences
```

```python
input_dir='./data/pathway_text/'
```

```python
pathway_list =[ os.path.join(input_dir, file_name) for file_name in os.listdir(input_dir)]
```

```python
pathway_list
```

```python
pathway_sentences_dict = {}
for pathway in pathway_list : 
    key = os.path.split(pathway)[1].split('.')[0]
    #print key
    if key in pathway_sentences_dict:
        print key
        raise Exception('there are same pathway')
    else: 
        #print pathway
        pathway_sentences_dict[key] = get_pathway_sentences(pathway)
    
```

```python
def get_sentence_list(sentence_string):
    senetence_list = [sentence+'.' for sentence in sentence_string.split('. ') if sentence != '']
    return senetence_list
```

```python
for key in pathway_sentences_dict.keys():
    pathway_sentences_dict[key] = get_sentence_list(pathway_sentences_dict[key])
```

```python
output_dir = './data/pathway_text_post_process/'
```

```python
for key in pathway_sentences_dict.keys():
    writer = open(output_dir+key+'.txt','w')
    writer.write('\n'.join(pathway_sentences_dict[key]))    
    writer.close()
```
